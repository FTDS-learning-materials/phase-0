## Matplotlib Subplot

In Matplotlib, subplots allow you to create multiple plots within a single figure. This is useful when you want to display multiple graphs or visualizations side by side for comparison or to present different aspects of your data.

```python
fig, ax = plt.subplots()
```

`fig, ax` means that while you running the `plt.subplots`, you also create the figure and axes.

```python
fig
```

fig basically only an image. So, if you apply the `.plot()` method onto fig, it's getting error.

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

fig.plot(x,y)
plt.show()
```

Meanwhile, `ax` is axes that is an `axes` instance so you can put on any methods to build your chart.

```python
ax
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

ax.plot(x,y)
fig
```

### Multiple subplots

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,3)) #ncols,nrows,figsize=(width,height)
```

Since you define a multiple subplot, the `axes` itself is a numpy array (it's like list but based on `numpy` library)

```python
axes #the axes consist of two elements
```

```python
x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]
```

```python
axes.plot(x1,x2)
```

If we run `axes.plot(x1,x2)`, it will be getting error because it's not `<Axes:>` instance but `numpy.ndarray` instance.

So, to insert a chart onto one of the axes, you have to access it by slicing the element like in list/tuple.

```python
axes[0].plot(x1,y1)
axes[1].plot(x2,y2)
fig
```

It's recommended to build charts step by step from defining fig and axes until you adding some features in a single cell.

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,3))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

plt.show()
```

## The Features

### Title

**Insert title onto each subplot**

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,3))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

#--- Define Title ---
axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

plt.show()
```

**Insert title to figure**

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,4))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

#--- Define Title to Each Subplot---
axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

#--- Define Figure's Title---
fig.suptitle("These are My Graphs")
fig.subplots_adjust(top=0.76) #Position setting from the suptitle

plt.show()
```

### Axis Label

```python
fig, axes = plt.subplots(ncols=2,figsize=(8,4))

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

axes[0].plot(x1,y1)
axes[1].plot(x2,y2)

axes[0].set_title("Exponential Curve")
axes[1].set_title("Linear Curve")

#Define axis label to the first subplot
axes[0].set_xlabel('X-Axis')
axes[0].set_ylabel('Y-Axis')

#Define axis label to the second subplot
axes[1].set_xlabel('X-Axis')
axes[1].set_ylabel('Y-Axis')

fig.suptitle("These are My Graphs")
fig.subplots_adjust(top=0.76)

plt.show()
```

### Legend

```python
fig, ax = plt.subplots()

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

ax.plot(x1,y1, label="exponential") #You have to set the label in every chart first
ax.plot(x2,y2, label="linear")

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.set_title('Our Curves')

plt.legend() #Using this code to show the legend
plt.show()
```

**Change the legend position**

```python
fig, ax = plt.subplots()

x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]

x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

ax.plot(x1,y1, label="exponential")
ax.plot(x2,y2, label="linear")

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.set_title('Our Curves')

plt.legend(bbox_to_anchor=(0.75,-0.15),ncol=2)

plt.show()
```

### Bubble Plot

For convinient, bubble lot can be built by `sns.scatterplot`. We can add colors, sizes, and markers to add more information/dimensions.

**Colors as Additional Info**

```python
fig, ax = plt.subplots()

ax.set_title('Heart Rate vs Daily Steps')
ax.set_xlabel('Heart Rate')
ax.set_ylabel('Daily Steps')

sns.scatterplot(data=data, x='Heart Rate',y='Daily Steps', hue='Gender', ax=ax)
plt.show()
```

**Markers as Additional Info**

```python
fig, ax = plt.subplots()

ax.set_title('Heart Rate vs Daily Steps')
ax.set_xlabel('Heart Rate')
ax.set_ylabel('Daily Steps')

sns.scatterplot(data=data, x='Heart Rate',y='Daily Steps', style='BMI Category', ax=ax)
plt.show()
```

**Sizes as Additional Info**

```python
fig, ax = plt.subplots()

ax.set_title('Heart Rate vs Daily Steps')
ax.set_xlabel('Heart Rate')
ax.set_ylabel('Daily Steps')

sns.scatterplot(data=data, x='Heart Rate',y='Daily Steps', size='Sleep Disorder', sizes=(10,100), ax=ax)
plt.show()
```

**Combining Hue, Style, and Size**

```python
fig, ax = plt.subplots()

ax.set_title('Heart Rate vs Daily Steps')
ax.set_xlabel('Heart Rate')
ax.set_ylabel('Daily Steps')

sns.scatterplot(data=data, x='Heart Rate',y='Daily Steps', hue='Gender', style='BMI Category', size='Sleep Disorder', ax=ax)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
